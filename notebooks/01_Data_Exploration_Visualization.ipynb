```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("../data/data set.csv")
df.shape
```

```python
df.info()
df["FLAG"].value_counts()
```

```python
print(df.columns[-10:])
```

```python
print("Total Missing Values:", df.isnull().sum().sum())
```

Duplicate

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
print(df["FLAG"].value_counts())

print("\nPercentage Distribution:")
print(df["FLAG"].value_counts(normalize=True)*100)
```

Split Honest and Theft Consumers

```python
honest = df[df["FLAG"] == 0]
theft = df[df["FLAG"] == 1]

print("Honest:", honest.shape)
print("Theft :", theft.shape)
```

Remove Non-Time-Series Columns

```python
honest_ts = honest.drop(columns=["CONS_NO","FLAG"])
theft_ts = theft.drop(columns=["CONS_NO","FLAG"])

print(honest_ts.shape)
print(theft_ts.shape)
```

```python
honest.to_csv("../data/honest_users.csv", index=False)
theft.to_csv("../data/theft_users.csv", index=False)
```

Visualizations

```python
plt.figure(figsize=(6,4))

df["FLAG"].value_counts().plot(kind="bar")

plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Consumers")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["FLAG"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=["Honest","Theft"]
)

plt.title("Percentage Distribution")
plt.ylabel("")

plt.show()
```

Missing Values Per Column (Top 20)

```python
missing = df.isnull().sum()

missing = missing[missing > 0]

missing = missing.sort_values(ascending=False).head(20)

plt.figure(figsize=(12,5))

missing.plot(kind="bar")

plt.title("Top 20 Columns with Missing Values")
plt.xlabel("Columns")
plt.ylabel("Missing Count")

plt.xticks(rotation=90)

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(15,6))

sns.heatmap(
    df.iloc[:200, :40].isnull(),
    cbar=False
)

plt.title("Missing Value Heatmap (First 200 Rows × First 40 Columns)")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df["1/1/2014"].hist(bins=30)

plt.title("Electricity Consumption on 2014/1/1")

plt.xlabel("Consumption")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.boxplot(df["1/1/2014"].dropna())

plt.title("Boxplot of Electricity Consumption")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    honest["1/1/2014"].dropna(),
    bins=30,
    alpha=0.6,
    label="Honest"
)

plt.hist(
    theft["1/1/2014"].dropna(),
    bins=30,
    alpha=0.6,
    label="Theft"
)

plt.legend()

plt.title("Consumption Comparison")

plt.xlabel("Consumption")

plt.ylabel("Frequency")

plt.show()
```

```python
sample = df.drop(columns=["CONS_NO","FLAG"]).iloc[:, :20]

corr = sample.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap (First 20 Days)")

plt.show()
```
